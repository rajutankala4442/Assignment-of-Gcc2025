#1. Basic Line Plot:
Write a Python program that uses Pyplot to create a simple line plot. Plot a set of x and y values and customize the appearance of the plot with labels, title, and grid lines.

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
#plotting
plt.plot(x,y)
#customizations
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('simple line plot')
plt.grid(True)
#display the plot
plt.show()
```

#2. Scatter Plot with Markers:
Write a Python program that uses Pyplot to create a scatter plot. Plot a set of x and y values with different markers for each data point. Customize the appearance of the plot with labels and a title.


```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
#plotting
plt.scatter(x,y,marker="o")
#customizations
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('scatter plot with markers')
plt.grid(True)
#display the plot
plt.show()
```

#3. Bar Plot with Labels:
Write a Python program that uses Pyplot to create a bar plot. Plot a set of x and y values as vertical bars with custom labels for each bar. Customize the appearance of the plot with labels and a title.

```python
import matplotlib.pyplot as plt
x=['A','B','C','D','E']
y=[10,7,5,8,12]
#plotting
plt.bar(x,y)
#customizations
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar plot with lables')
#display the plot
plt.show()
```

#4. Histogram Plot:
Write a Python program that uses Pyplot to create a histogram plot. Generate random data and plot a histogram to visualize the distribution. Customize the appearance of the plot with labels and a title.


```python
import matplotlib.pyplot as plt
import random
#random data
data=[random.randint(1,100)for _ in range(100)]
#plotting
plt.hist(data,bins=10)
#customizations
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Histogram Plot')
#display the plot
plt.show()
```

#5. Pie Chart:
Write a Python program that uses Pyplot to create a pie chart. Plot a set of values as a pie chart, with labels and percentages for each slice. Customize the appearance of the chart with labels and a title.

```python
import matplotlib.pyplot as plt
# Data
x=['A','B','C','D','E']
y=[15,30,25,10,20]
#plotting
plt.pie(y,labels=x,autopct='%1.1f%%')
#customizations
plt.title('Pie Chart')
#display the plot
plt.show()
```
